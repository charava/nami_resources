```python
import pandas as pd
import pdfquery
import re
```

```python
# Read the text file
with open('octyouthguide.txt', 'r') as file:
    text = file.read()


# Split the text into sections using the delimiter "________________"
sections = text.split('________________')

# Save each section into separate variables
section1, section2, section3, section4, section5, section6, section7, section8, section9, section10 = [section.strip() for section in sections]
SECTIONS = [section1, section2, section3, section4, section5, section6, section7, section8, section9, section10]
```

```python
# THIS IS TO GET ALL THE SECTIONS
SECTION_TITLES = []
for ind in range(10):
    
    section_title = SECTIONS[ind].split('\n', 1)[0]
    section_title_cleaned = re.sub(r'[^a-zA-Z\s]', '', section_title)

    SECTION_TITLES.append(section_title_cleaned)
```

```python
SECTION_TITLES = ['Community Clinics',
 'Crisis/Evaluation Services',
 'Education Services',
 'Extended Services',
 'FosterAdoptiveAlternative Family Resources',
 'General InformationResourcesServices',
 'Homeless Resources',
 'LGBTQ Centered Services',
 'Support Groups',
 'Outpatient Therapy']
```

```python
SERVICE_NAME = []
SUMMARY = []
FEES_PROGRAMS = []
LANGUAGES=[]
WEBSITE = []
ADDRESS = []
EMAIL =[]
PHONE= []
CATEGORY = []

for section in range(len(SECTIONS)):


    section_split = [fragment.strip() for fragment in re.split(r'\t|\n', SECTIONS[section])]


    # Print the fragments
    section_split_list = []
    for fragment in section_split:
        if not (fragment == ''):
            section_split_list.append(fragment)

    section_split_list = section_split_list[1:] # taking out the first item cuz its the title of the whole section

    

    match = re.search(r'(?<=:\s)(.*)', text)

    for entry in section_split_list:
        if 'Summary' in entry:
            match = re.search(r'(?<=:\s)(.*)', entry)
            if match:
                address = match.group(1)
            SUMMARY.append(address)

        elif 'Fees/Programs' in entry:
            match = re.search(r'(?<=:\s)(.*)', entry)
            if match:
                address = match.group(1)
            FEES_PROGRAMS.append(address)
        elif 'Language' in entry:
            match = re.search(r'(?<=:\s)(.*)', entry)
            if match:
                address = match.group(1)
            LANGUAGES.append(address)
        elif 'Website' in entry:
            match = re.search(r'(?<=:\s)(.*)', entry)
            if match:
                address = match.group(1)
            WEBSITE.append(address)
        elif 'Address' in entry:
            match = re.search(r'(?<=:\s)(.*)', entry)
            if match:
                address = match.group(1)
            ADDRESS.append(address)
        elif 'Email' in entry:
            match = re.search(r'(?<=:\s)(.*)', entry)
            if match:
                address = match.group(1)
            EMAIL.append(address)
        elif 'Phone' in entry:
            match = re.search(r'(?<=:\s)(.*)', entry)
            if match:
                address = match.group(1)
            PHONE.append(address)
        else:
            SERVICE_NAME.append(entry)


    
    for i in range(len(FEES_PROGRAMS)):
        CATEGORY.append(SECTION_TITLES[section])

df = pd.DataFrame([SERVICE_NAME, CATEGORY, SUMMARY, FEES_PROGRAMS, LANGUAGES, WEBSITE, ADDRESS, EMAIL, PHONE])
df =df.T
newname = 'combined.csv'
df.to_csv(newname)

```
